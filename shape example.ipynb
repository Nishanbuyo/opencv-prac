```python
import numpy as np
import cv2 as cv
drawing = False # true if mouse is pressed
mode = True # if True, draw rectangle. Press 'm' to toggle to curve
ix,iy = -1,-1
# mouse callback function
def nothing(x):
    pass
def draw_circle(event,x,y,flags,param):
    global ix,iy,drawing,mode
    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv.rectangle(img,(ix,iy),(x,y),(r,g,b),-1)
            else:
                cv.circle(img,(x,y),5,(r,g,b),-1)
    elif event == cv.EVENT_LBUTTONUP:
        drawing = False
        if mode == True:
            cv.rectangle(img,(ix,iy),(x,y),(r,g,b),-1)
        else:
            cv.circle(img,(x,y),5,(r,g,b),-1)
            
img = np.zeros((512,512,3), np.uint8)
cv.namedWindow('image')
cv.createTrackbar('R','image',0,255,nothing)
cv.createTrackbar('G','image',0,255,nothing)
cv.createTrackbar('B','image',0,255,nothing)
cv.setMouseCallback('image',draw_circle)
while(1):
    cv.imshow('image',img)
    k = cv.waitKey(1) & 0xFF
    if k == ord('m'):
        mode = not mode
    elif k == 27:
        break
    r = cv.getTrackbarPos('R','image')
    g = cv.getTrackbarPos('G','image')
    b = cv.getTrackbarPos('B','image')
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
def nothing(x):
    pass
# Create a black image, a window
img = np.zeros((300,512,3), np.uint8)
cv.namedWindow('image')
# create trackbars for color change
cv.createTrackbar('R','image',0,255,nothing)
cv.createTrackbar('G','image',0,255,nothing)
cv.createTrackbar('B','image',0,255,nothing)
# create switch for ON/OFF functionality
switch = '0 : OFF \n1 : ON'
cv.createTrackbar(switch, 'image',0,1,nothing)
while(1):
    cv.imshow('image',img)
    k = cv.waitKey(1) & 0xFF
    if k == 27:
        break
    # get current positions of four trackbars
    r = cv.getTrackbarPos('R','image')
    g = cv.getTrackbarPos('G','image')
    b = cv.getTrackbarPos('B','image')
    s = cv.getTrackbarPos(switch,'image')
    if s == 0:
        img[:] = 0
    else:
        img[:] = [b,g,r]
cv.destroyAllWindows()

```

```python
import cv2 as cv
url='http://192.168.1.2:8080/video'
cam=cv.VideoCapture(url)


def sketch_transform(image):
    gray_image= cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    gray_image_blur = cv.GaussianBlur(gray_image, (7,7), 0 )
    image_canny= cv.Canny(gray_image_blur, 10, 80)
    _, mask = cv.threshold(image_canny, 30, 255, cv.THRESH_BINARY_INV)
    return mask
while True:
    _, frame = cam.read()
    showCrosshair = False
    fromCenter = False
    r=cv.selectROI("Image", frame, showCrosshair, fromCenter)
    break
    
while True:
    _, image_frame= cam.read()
    rect_img = image_frame[int(r[1]): int(r[1] + r[3]), int(r[0]): int(r[0]+r[2])]
    sketch_rect =sketch_transform(rect_img)
    sketch_rect_rgb =cv.cvtColor(sketch_rect, cv.COLOR_GRAY2BGR)
    image_frame[int(r[1]): int(r[1] + r[3]), int(r[0]): int(r[0]+r[2])]= sketch_rect_rgb
    cv.imshow("Sketcher Roi", image_frame)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
        
cam.release()
cv.destroyAllWindows()
    
```
